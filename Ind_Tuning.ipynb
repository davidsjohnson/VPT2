```python
p6_r.loc[p6_r['param_SVC__C']==.11, 'param_SVC__C'] = .1
```

```python
svc_cs = [.01, .1, 10, 100]
svc_gammas = [.001, .01, .1, 1, np.nan]
```

```python
plt.figure(figsize=(12,8))

for i, df in enumerate([p1_r, p3_r, p4_r, p6_r]):
    plt.subplot(2, 2, i+1)
    ax = plt.gca()
    ax.set_xscale("log", nonposx='clip')
    plt.ylim([.8, 1.0])
    plt.yticks(np.linspace(.8, 1, 5))
    for i, gamma in enumerate(svc_gammas):

        if gamma in [.001, .01, .1, 1, 10]:
            p = df.loc[df['param_SVC__gamma'] == gamma]
            label = gamma
        else:
            p = df.loc[~df['param_SVC__gamma'].isin([.001, .01, .1, 1, 10])]
            label = "Linear SVC"
        plt.plot(svc_cs, p["mean_test_accuracy"], label=label)
    plt.legend(title='Gammas')
```
